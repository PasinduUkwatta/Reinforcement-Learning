```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
y=np.arange(0,30,3)
x,y
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,10)
plt.ylim(0,30)
plt.title("String Title")
plt.savefig("my_first_plot.png")
```

```python
 a =np.linspace(0,10,11)
```

```python
b=a**4
a,b
```

```python
x=np.arange(0,10)
y=2*x

x,y
```

```python
fig =plt.figure()

#large axes
axes1 =fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

#small axes
axes2=fig.add_axes([0.1,0.4,0.5,0.5])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")
plt.savefig('double_figures.png')
```

```python
fig =plt.figure(dpi=200)

axes1 =fig.add_axes([0,0,1,1])
axes1.plot(a,b)
plt.savefig('figures.png')
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2,dpi=150)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout()
plt.subplots_adjust(wspace=.8)

# save the figure
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
```
