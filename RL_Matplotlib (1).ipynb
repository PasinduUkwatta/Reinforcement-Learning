```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
y=np.arange(0,30,3)
x,y
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,10)
plt.ylim(0,30)
plt.title("String Title")
plt.savefig("my_first_plot.png")
```

```python
 a =np.linspace(0,10,11)
```

```python
b=a**4
a,b
```

```python
x=np.arange(0,10)
y=2*x

x,y
```

```python
fig =plt.figure()

#large axes
axes1 =fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

#small axes
axes2=fig.add_axes([0.1,0.4,0.5,0.5])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")
plt.savefig('double_figures.png')
```

```python
fig =plt.figure(dpi=200)

axes1 =fig.add_axes([0,0,1,1])
axes1.plot(a,b)
plt.savefig('figures.png')
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2,dpi=150)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout()
plt.subplots_adjust(wspace=.8)

# save the figure
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = [0, 2, 4, 6]
y = [1, 3, 4, 8]


plt.plot(x,y)


plt.xlabel('x values')
plt.ylabel('y values')
plt.title('plotted x and y values')
plt.legend(['line 1'])


# save the figure
plt.savefig('plot.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
plt.plot(x,(x),label="X VS X")
plt.legend()
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x,color="#1ce00d")#rgb hex code
y=np.power(x,2)
ax.plot(x,y,color="#de1895")#rgb hex code
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x,color="#1ce00d",lw=3)#rgb hex code
y=np.power(x,2)
ax.plot(x,y,color="#de1895",lw=10)#rgb hex code
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,x,color="#1ce00d",lw=3)#rgb hex code
y=np.power(x,2)
ax.plot(x,y,color="#de1895",lw=2,marker="o",ls="--")#rgb hex code
```
